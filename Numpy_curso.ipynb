```python
# Importa a biblioteca NumPy para operações numéricas.
import numpy as np
```

```python
# Define a URL do arquivo CSV contendo os dados.
url = 'https://raw.githubusercontent.com/alura-cursos/numpy/dados/apples_ts.csv'
```

```python
# Carrega os dados da URL em um array NumPy 'dado'.
# Os dados são delimitados por vírgulas e as colunas de 1 a 87 (exclusivo) são usadas.
dado = np.loadtxt(url, delimiter=',', usecols=np.arange(1,88,1))
```

```python
# Verifica o número de dimensões (rank) do array 'dado'.
dado.ndim
```

```python
# Verifica o número total de elementos no array 'dado'.
dado.size
```

```python
# Verifica a forma (dimensões) do array 'dado'.
dado.shape
```

```python
# Transpõe o array 'dado'. Isso inverte as linhas e colunas.
dado_transposto = dado.T
```

```python
# Extrai a primeira coluna (índice 0) do array transposto para 'datas'.
# Esta coluna provavelmente representa datas ou passos de tempo.
datas = dado_transposto[:,0]
```

```python
# Extrai as colunas de 1 a 5 (excluindo a 6) do array transposto para 'precos'.
# Isso provavelmente representa dados de preços para diferentes regiões/categorias.
precos = dado_transposto[:,1:6]
```

```python
# Importa a biblioteca Matplotlib para plotagem, especificamente o módulo pyplot como 'plt'.
import matplotlib.pyplot as plt
```

```python
# Reatribui 'datas' a um novo array NumPy representando uma sequência de 1 a 87.
# Isso sobrescreve a variável 'datas' anterior, que continha dados do CSV.
datas = np.arange(1,88,1)
```

```python
# Plota o array 'datas' (eixo x) contra a primeira coluna de 'precos' (eixo y).
# Isso visualiza a tendência de preços para a primeira categoria/região ao longo do tempo.
plt.plot(datas, precos[:,0])
```

```python
# Atribui colunas específicas do array 'precos' a variáveis individuais de cidades.
# Cada variável agora contém os dados de preço para uma cidade específica.
Moscow = precos[:,0]
Kaliningrad = precos[:,1]
Petersburg = precos[:,2]
Krasnodar = precos[:,3]
Ekaterinburg = precos[:,4]
```

```python
# Atribui o array 'datas' a uma nova variável 'x' para uso em uma equação linear.
x = datas
```

```python
# Define uma equação linear simples com 'x' como a variável independente.
y = 0.52 * x + 80
```

```python
# Calcula a norma euclidiana (norma L2) entre os dados de 'Moscow' e a equação linear 'y'.
# Isso mede a 'distância' ou diferença entre os dados reais e o modelo linear.
np.linalg.norm(Moscow - y)
```

```python
# Plota os dados originais de 'Moscow' e a equação linear 'y' no mesmo gráfico.
# Isso visualiza quão bem o modelo linear se ajusta aos dados de Moscou.
plt.plot(datas, Moscow)
plt.plot(x,y)
```

```python
# Atribui os dados de 'Moscow' a 'Y' e 'datas' a 'X' para cálculos de regressão linear.
# 'n' é definido como o número de elementos em 'Moscow', representando o número de pontos de dados.
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
# Verifica a forma do array resultante do quadrado de cada elemento de 'X'.
(X**2).shape
```

```python
# Calcula o coeficiente 'a' (inclinação) para um modelo de regressão linear simples
# usando o método dos mínimos quadrados.
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
```

image.png

```python
# Calcula o coeficiente 'b' (intercepto y) para o modelo de regressão linear
# usando a inclinação 'a' calculada e as médias de 'X' e 'Y'.
b = np.mean(Y) - a*np.mean(X)
```

```python
# Define a linha de regressão linear 'y' usando os coeficientes 'a' e 'b' calculados.
y = a*X+b
```

```python
# Calcula a norma euclidiana (norma L2) entre os dados de 'Moscow' e a linha de regressão linear 'y'.
# Isso indica o ajuste da linha de regressão aos dados reais.
np.linalg.norm(Moscow-y)
```

```python
# Plota os dados originais de 'Moscow' e a linha de regressão linear 'y' calculada.
plt.plot(datas, Moscow)
plt.plot(x,y)
```

```python
# Plota os dados de 'Moscow', a linha de regressão linear e um ponto específico (41.5, 41.5*a+b).
# O ponto é marcado com um asterisco vermelho.
plt.plot(datas,Moscow)
plt.plot(x,y)
plt.plot(41.5,41.5*a+b,'*r')
```

```python
# Plota os dados de 'Moscow', a linha de regressão linear e dois pontos específicos.
# O primeiro ponto (41.5, 41.5*a+b) faz parte da linha, e o segundo ponto (100, 100*a+b) é uma previsão/extrapolação,
# marcado com um asterisco vermelho.
plt.plot(datas,Moscow)
plt.plot(x,y)
plt.plot(41.5,41.5*a+b)
plt.plot(100,100*a+b,'*r')
```

```python
# Gera um array de 100 inteiros aleatórios entre 40 (inclusive) e 100 (exclusivo).
np.random.randint(low=40,high=100,size=100)
```

```python
# Gera um array de 100 números de ponto flutuante aleatórios (coeficientes) uniformemente distribuídos
# entre 0.10 (inclusive) e 0.90 (exclusivo).
coef_angulares = np.random.uniform(low=0.10,high=0.90,size=100)
```

```python
# Inicializa um array NumPy vazio 'norma2'.
# Itera 100 vezes, calculando a norma euclidiana entre os dados de 'Moscow' e um modelo linear
# onde a inclinação é retirada de 'coef_angulares' e o intercepto 'b' é constante.
# Cada norma calculada é anexada ao array 'norma2'.
norma2 = np.array([])
for i in range(100):
    norma2 = np.append(norma2, np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
```

```python
# Exibe o array 'norma2', que contém as normas euclidianas para 100 modelos lineares diferentes.
norma2
```

```python
# Exibe o segundo elemento (índice 1) do array 'coef_angulares'.
coef_angulares[1]
```

```python
# Define a semente aleatória para reprodutibilidade da geração de números aleatórios.
np.random.seed(84)
# Gera um novo array de 100 números de ponto flutuante aleatórios para coeficientes.
coef_angulares = np.random.uniform(low=0.10,high=0.90,size=100)
# Inicializa um array vazio para as normas.
norma2 = np.array([])
# Recalcula as normas euclidianas para o novo conjunto de coeficientes e as anexa a 'norma2'.
for i in range(100):
    norma2 = np.append(norma2, np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
```

```python
# Empilha o array 'norma2' (normas euclidianas) e o array 'coef_angulares' (inclinações) lado a lado.
# Isso cria um novo array 2D 'dados' onde cada linha contém uma norma e seu coeficiente correspondente.
dados = np.column_stack([norma2,coef_angulares])
```

```python
# Verifica a forma do array 'dados'. Deve ser (100, 2).
dados.shape
```

```python
# Salva o array 'dados' em um arquivo CSV chamado 'dados.csv', usando vírgula como delimitador.
np.savetxt('dados.csv',dados,delimiter=',')
```
